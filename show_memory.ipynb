Die roten und grünen marker representiert das überschreiten des positiven oder negativen schwellenwerts, und nicht kaufen und verkaufen.

The red and green markers represent exceeding the positive or negative threshold, and not buying and selling. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec 
%matplotlib inline
```

```python
SENTIMENT_THRESHOLD = 0.001
NEGATIVE_SENTIMENT_THRESHOLD = -0.001
PAIRING = 'USDT'

coin_dic = {
    'XRP': ['ripple', 'xrp', 'XRP', 'Ripple', 'RIPPLE'],
    'BTC': ['BTC', 'bitcoin', 'Bitcoin', 'BITCOIN'],
    'ETH': ['ETH', 'Ethereum'],
    'BNB' : ['BNB', 'Binance Coin'],
    'LTC': ['LTC', 'Litecoin']
    }
```

```python
df = pd.read_csv("memory.csv")
```

```python
num = len(coin_dic)

fig = plt.figure(figsize =(16,num*4))
gs1 = gridspec.GridSpec(num, 1, figure=fig)
axs = []

for coin in coin_dic:
    axs.append(fig.add_subplot(gs1[num-1]))    
    axs[-1].plot(df[coin+PAIRING + '-price'],'-b', label=coin+PAIRING + '-price')
    axs[-1].plot(df[coin+PAIRING + '-price'][df[coin+PAIRING + '-sentiment']>SENTIMENT_THRESHOLD],'^', markersize=8, color='g', label='Pos')
    axs[-1].plot(df[coin+PAIRING + '-price'][df[coin+PAIRING + '-sentiment']<NEGATIVE_SENTIMENT_THRESHOLD],'v', markersize=8, color='r', label ='Neg')
    axs[-1].set_xlabel('Time')
    axs[-1].set_ylabel('Price')
    axs[-1].legend()    
    num = num - 1
```
